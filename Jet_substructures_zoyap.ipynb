```python
import numpy as np
import awkward as ak
import matplotlib.pyplot as plt
import mplhep as hep
import hist
from coffea.nanoevents import NanoEventsFactory
from coffea.nanoevents.schemas.nanoaod import NanoAODSchema 
import vector as vec

plt.style.use(hep.style.CMS) 
NanoAODSchema.warn_missing_crossrefs = False
```

```python
BASE_PATH = "root://cmsxrootd.fnal.gov//store/user/cmsdas/2025/Jets/RunIISummer20UL18NanoAODv9/"


fname_qcd = BASE_PATH + "QCD_Pt-15to7000_TuneCP5_flat2018_13TeV-pythia8/NANOAODSIM/20UL18JMENano_106X_upgrade2018_realistic_v16_L1v1/2520000/0179D19E-E662-A845-8B92-F9F63702A884.root" + ":Events"
fname_ttbar = BASE_PATH + "TTJets_TuneCP5_13TeV-amcatnloFXFX-pythia8/NANOAODSIM/20UL18JMENano_106X_upgrade2018_realistic_v16_L1v1-v1/2520000/47A0D984-2D6F-0D42-B5CB-906786434349.root" + ":Events"
fname_rsgluon = BASE_PATH + "RSGluonToTT_M-1000_TuneCP5_13TeV-amcatnloFXFX-pythia8/NANOAODSIM/20UL18JMENano_106X_upgrade2018_realistic_v16_L1v1/2520000/003C891C-931B-304D-9345-56B98D5804DF.root" + ":Events"
fname_wjets = BASE_PATH + "WJetsToQQ_HT-800toInf_TuneCP5_13TeV-madgraphMLM-pythia8/NANOAODSIM/20UL18JMENano_106X_upgrade2018_realistic_v16_L1v1/2520000/01D2197A-A905-BE46-8153-6D6D90C28A18.root" + ":Events"


print("Loading all datasets...")
events_qcd = NanoEventsFactory.from_root(fname_qcd, schemaclass=NanoAODSchema).events() 
events_ttbar = NanoEventsFactory.from_root(fname_ttbar, schemaclass=NanoAODSchema).events()
events_rsgluon = NanoEventsFactory.from_root(fname_rsgluon, schemaclass=NanoAODSchema).events()
events_wjets = NanoEventsFactory.from_root(fname_wjets, schemaclass=NanoAODSchema).events()
print("Data loading complete. Proceeding to plotting.")

```

```python
# Ex. 4.1
fig, ax = plt.subplots()

hists_mass = (
    hist.Hist(
        hist.axis.StrCategory(["QCD", "ttbar", "RSGluon", "Wjets"], name="dataset", label="Dataset"),
        hist.axis.Regular(50, 50, 300, name="mass", label="FatJet softdrop mass [GeV]"),
    )
    .fill( "QCD", ak.flatten(events_qcd.FatJet.msoftdrop) )
    .fill( "ttbar", ak.flatten(events_ttbar.FatJet.msoftdrop) )
    .fill( "RSGluon", ak.flatten(events_rsgluon.FatJet.msoftdrop) )
    .fill( "Wjets", ak.flatten(events_wjets.FatJet.msoftdrop) )
)

hists_mass.plot1d(ax=ax, density=True)
ax.legend()
ax.set_yscale('log')
ax.set_title("Exercise 4.1: Jet Mass Distribution")
plt.show()
```

```python
# Ex 4.2
hists_grooming = (
    hist.Hist(
        hist.axis.StrCategory(["mass", "msoftdrop"], name="mass_type", label="Mass Type"),
        hist.axis.Regular(200, 0, 200, name="mass", label="Jet Mass [GeV]"),
    )
    .fill( "mass", ak.flatten(events_ttbar.FatJet.mass) )
    .fill( "msoftdrop", ak.flatten(events_ttbar.FatJet.msoftdrop) )
)

fig, ax = plt.subplots()
hists_grooming.plot1d(ax=ax, density=Tue)
ax.legend()
ax.set_title("Exercise 4.2: Ungroomed vs. Softdrop Mass (ttbar)")
plt.show()
```

```python
# Rho parameter
fig, ax = plt.subplots()

hists_rho = (
    hist.Hist(
        hist.axis.StrCategory(["QCD", "ttbar", "RSGluon", "Wjets"], name="dataset", label="Dataset"),
        hist.axis.Regular(20, -10, 0, name="rho", label=r"$\rho$ parameter"),
    )
    .fill( "QCD", ak.flatten(np.log(np.power(events_qcd.FatJet.msoftdrop / (events_qcd.FatJet.pt * 0.8), 2))) )
    .fill( "ttbar", ak.flatten(np.log(np.power(events_ttbar.FatJet.msoftdrop / (events_ttbar.FatJet.pt * 0.8), 2))) )
    .fill( "RSGluon", ak.flatten(np.log(np.power(events_rsgluon.FatJet.msoftdrop / (events_rsgluon.FatJet.pt * 0.8), 2))) )
    .fill( "Wjets", ak.flatten(np.log(np.power(events_wjets.FatJet.msoftdrop / (events_wjets.FatJet.pt * 0.8), 2))) )
)

hists_rho.plot1d(ax=ax, density=True)
ax.legend()
ax.set_title(r"$\rho$ Parameter Distribution")
plt.show()
```
